```python
import requests
import os
from bs4 import BeautifulSoup
URL_busqueda = "https://cafeytableros.cl/shop/?product-page=1"
filename = "cafeytableroscl.txt"
if os.path.exists(filename):
    os.remove(filename)
for i in range(0, 100):
    URL = URL_busqueda
    # print(URL)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    pagina = soup.find(id='main-content')
    juegos = pagina.find_all("div", class_="entry-content")
    if not juegos:
        break
    listado = juegos[0].find("ul",
                             class_="products columns-3").text.replace("¡Oferta!", "").replace(" pesos", "").split("\n")
    while '' in listado:
        listado.remove('')
    for j in range(0, len(listado), 2):
        nombre = listado[j]
        nombre = nombre.replace('Ampliación 5-6 Jugadores',"5 6 Jugadores")
        nombre = nombre.replace('(Blister)',"").replace('Blister',"")
        nombre = nombre.replace('–',"")
        nombre = nombre.strip().title()
        precio = listado[j + 1].split(" ")[-1].replace("$", "").replace(",", "")
        with open(filename, 'a', encoding='utf-8') as f:
            f.write(nombre + " " + precio + "\n")
    if len(listado) < 18:
        break
    pag_actual = pagina.find_all("span", class_="page-numbers current")
    pag_actual = int(pag_actual[0].text)
    pag_siguiente = pag_actual + 1
    URL_busqueda = URL.replace("page=%d" % pag_actual, "page=%d" % pag_siguiente)
text = open(filename)
lines = text.readlines()
lines.sort()
text.close()
with open(filename, 'w') as f:
    for line in lines:
        f.write(line)
```
