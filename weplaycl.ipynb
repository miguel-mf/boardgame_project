```python
import requests
import os
from bs4 import BeautifulSoup
URL_busqueda = "https://www.weplay.cl/juegos-de-mesa-y-cartas/juegos-de-mesa.html?p=1"
filename="weplaycl.txt"
if os.path.exists(filename):
    os.remove(filename)
for i in range(0,100):
    URL = URL_busqueda
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    pagina = soup.find(id='layer-product-list')
    juegos = pagina.find_all("div", class_="product-item-info type1")
    if not juegos:
        break
    for juego in juegos:
        nombre = juego.find("strong", class_="product name product-item-name")
        nombre = nombre.text.replace("\n","").strip().lower()
        nombre = nombre.replace("base","")
        nombre = nombre.replace("ampliación 5 – 6 jugadores","5 6 jugadores")
        nombre = nombre.replace("ampl 5-6 jug.","5 6 jugadores")
        nombre = nombre.replace("ampliacion 5-6 jug.","5 6 jugadores")
        nombre = nombre.replace("juego de mesa :", "").replace("juego de mesa:", "").replace("juego de mesa","")
        nombre = nombre.replace("juego de cartas :", "").replace("juego de cartas:", "").replace("juego de cartas", "")
        nombre = nombre.strip().title()
        if nombre[0:2] == '- ':
          nombre = nombre[2:-1]
        aux_precio = juego.find("div", class_="price-box price-final_price")
        precio = aux_precio.find("span", class_="price-wrapper").text.strip().replace("$","").replace(".","")
        with open(filename, 'a', encoding='utf-8') as f:
            f.write(nombre + " " + precio + "\n")
    pag_actual = pagina.find_all("li", class_="item current")
    pag_actual = int(pag_actual[0].text.strip().split("\n")[1])
    pag_siguiente = pag_actual+1
    URL_busqueda = URL.replace("p=%d" % pag_actual,"p=%d" % pag_siguiente)
text = open(filename)
lines = text.readlines()
lines.sort()
text.close()
with open(filename, 'w') as f:
    for line in lines:
        f.write(line)
```
